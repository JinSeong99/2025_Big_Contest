```python
from sklearn.preprocessing import MinMaxScaler

# 동일 업종 매출금액 비율, 동일 업종 매출건수 비율 전처리 -> 추후 지표 설계를 위함
# 극단값과 평균 분포 차이가 심하고 많음 -> 하지만 제거하기에는 해당 데이터가 가지고 있는 순서 정보가 왜곡 될 수 있음

# Winsorization 함수 정의 (상하위 분위수로 극단값 잘라내기)
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# Winsorization 함수 정의
def winsorize(series, lower=0.01, upper=0.99):
    low, high = series.quantile(lower), series.quantile(upper)
    return series.clip(lower=low, upper=high)

cols = ["동일 업종 매출금액 비율", "동일 업종 매출건수 비율"]

scaled_data = {}

# 1. Winsorization → 로그 변환 → MinMax Scaling
for c in cols:
    # Winsorization
    win = winsorize(data_set2[c])
    
    # 로그 변환
    loged = np.log1p(win)
    
    # MinMax Scaling
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(loged.values.reshape(-1, 1)).flatten()
    
    # 최종 결과 저장
    scaled_data[c + "_scaled"] = scaled

# 2. DataFrame 변환
scaled_df = pd.DataFrame(scaled_data, index=data_set2.index)

# 3. 최종 병합
data_set2 = pd.concat([data_set2, scaled_df], axis=1)

# 확인
print(data_set2[[c + "_scaled" for c in cols]].head())


```

```python
from sklearn.preprocessing import RobustScaler, QuantileTransformer
import matplotlib.pyplot as plt
import seaborn as sns

# 타겟 변수에 기반한 전처리 방법 비교 
# 1. 폐업여부 컬럼 생성
data_set1["폐업여부"] = data_set1["폐업일"].notna().astype(int)

# 2. 데이터 병합
merged = pd.merge(
    data_set2,
    data_set1[["가맹점구분번호", "폐업여부"]],
    on="가맹점구분번호",
    how="left"
)

# 3. 순위 변수
rank_cols = ["동일 업종 내 매출 순위 비율", "동일 상권 내 매출 순위 비율"]

# 전처리기 준비
rob_scaler = RobustScaler()
qt = QuantileTransformer(output_distribution="uniform", random_state=42)

# 4. 시각화 (각 변수별 2가지 방식 비교 - 컬럼 추가 X)
for col in rank_cols:
    # Robust Scaling (임시)
    robust_vals = rob_scaler.fit_transform(merged[[col]])
    
    # Percentile → Robust (임시)
    pct_vals = qt.fit_transform(merged[[col]])
    pct_robust_vals = rob_scaler.fit_transform(pct_vals)
    
    # 시각화
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))

    # (a) Robust Scaling - Boxplot
    sns.boxplot(x=merged["폐업여부"], y=robust_vals.flatten(), ax=axes[0,0], palette="Set2")
    axes[0,0].set_title(f"{col} (Robust) - 폐업 vs 생존")

    # (b) Robust Scaling - Histogram
    sns.histplot(x=robust_vals.flatten(), hue=merged["폐업여부"], kde=True,
                 element="step", stat="density", common_norm=False, ax=axes[0,1])
    axes[0,1].set_title(f"{col} (Robust) 분포")

    # (c) Percentile+Robust - Boxplot
    sns.boxplot(x=merged["폐업여부"], y=pct_robust_vals.flatten(), ax=axes[1,0], palette="Set2")
    axes[1,0].set_title(f"{col} (Percentile+Robust) - 폐업 vs 생존")

    # (d) Percentile+Robust - Histogram
    sns.histplot(x=pct_robust_vals.flatten(), hue=merged["폐업여부"], kde=True,
                 element="step", stat="density", common_norm=False, ax=axes[1,1])
    axes[1,1].set_title(f"{col}  분포")

    plt.tight_layout()
    plt.show()

# 극간값표현이 명확하게 나왔고 -1~1 까지 고르게 표현됨(동일상권 히스토) 다른 그래프도 마찬가지
# robust 방법에 이상치가 많음
# 인사이트: 동일 업종, 상권 내 매출 순위 가 높은데 폐업을 더많이 함

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler
from scipy.stats.mstats import winsorize

# 전처리 대상 : "동일 업종 내 해지 가맹점 비중", "동일 상권 내 해지 가맹점 비중" (0-100)
target_cols = ["동일 업종 내 해지 가맹점 비중", "동일 상권 내 해지 가맹점 비중"]

# 1. Winsorization + Robust Scaling 적용
processed_data = {}
for col in target_cols:
    arr = data_set2[col].to_numpy()
    
    # Winsorization (상하위 1% 극단치 잘라냄 → 필요에 따라 limits 조정)
    winsorized = winsorize(arr, limits=[0.01, 0.01])  
    
    # Robust Scaling
    minmax_scaler = MinMaxScaler()
    scaled = minmax_scaler.fit_transform(winsorized.reshape(-1, 1)).flatten()

    
    # 최종 컬럼명 저장
    processed_data[col + "_scaled"] = scaled

# 2. 최종 결과 DataFrame 생성
final_scaled_df = pd.DataFrame(processed_data, index=data_set2.index)

# 3. 기존 data_set2에 병합
data_set2 = pd.concat([data_set2, final_scaled_df], axis=1)

# 확인
print(data_set2[[c for c in data_set2.columns if "_scaled" in c]].head())

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================
# 1️⃣ 데이터 준비
# ==========================
closed_df = last_df[last_df['폐업여부'] == 1].copy()

# 폐업시점(최대 기준년월)을 기준으로 시점  마지막으로 관측된 시점
closed_df['폐업시점'] = closed_df.groupby('가맹점구분번호')['기준년월'].transform('max')
closed_df['폐업전개월'] = closed_df['기준년월'] - closed_df['폐업시점']  # ex. -12 ~ 0

# 분석 지표
indicators = ['매출안정성지표', '경쟁우위 지표', '고객 충성도 지표']

# 폐업 전 12개월만
closed_df = closed_df[closed_df['폐업전개월'].between(-12, 0)]

# ==========================
# 2️⃣ 지표별 평균 추세
# ==========================
trend = (
    closed_df.groupby('폐업전개월')[indicators]
    .mean()
    .reset_index()
    .sort_values('폐업전개월')
)

# ==========================
# 3️⃣ 시각화
# ==========================
plt.figure(figsize=(10,6))
colors = ['#FF6F61', '#6B5B95', '#88B04B']
for c, col in zip(colors, indicators):
    sns.lineplot(data=trend, x='폐업전개월', y=col, label=col, color=c)

plt.axvline(x=-6, color='red', linestyle='--', label='폐업 6개월 전')
plt.title('폐업 가맹점 주요 지표 평균 추세 (폐업 기준 정렬)')
plt.xlabel('폐업 전 개월수 (0=폐업월)')
plt.ylabel('지표 평균값')
plt.legend()
plt.show()
```

```python
# # =====================================
# 지표별 경고/위험 임계치 계산 따로 변수 저장
# =====================================

# 각 지표별로 구간별 평균 계산
threshold_dict = {}

for col in indicators:
    # 정상(안정 구간): -12 ~ -6개월
    normal_mean = trend.loc[trend['폐업전개월'].between(-12, -6), col].mean()
    # 경고(조기 위험): -6 ~ -4개월
    warning_mean = trend.loc[trend['폐업전개월'].between(-6, -4), col].mean()
    # 위험(급락 구간): -4 ~ 0개월
    danger_mean = trend.loc[trend['폐업전개월'].between(-4, 0), col].mean()

    threshold_dict[col] = {
        '정상기준': round(normal_mean, 3),
        '경고임계치': round(warning_mean, 3),
        '위험임계치': round(danger_mean, 3)
    }

# 임계치 DataFrame으로 변환
threshold_df = pd.DataFrame(threshold_dict).T
threshold_df.index.name = '지표명'

# 결과 확인
print("\n📊 지표별 경고/위험 임계치 요약:")
display(threshold_df)
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================
#  평가 지표 함수
# ==========================
def evaluate_forecast(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / np.clip(np.abs(y_true), 1e-8, None))) * 100
    return mae, rmse, mape


# ==========================
#  Prophet 기반 시계열 예측 함수 (last_df 구조 대응)
# ==========================
def evaluate_forecast_model_prophet(last_df, threshold_df, forecast_months=10, pre_close_months=6):
    """
    Prophet 기반 시계열 예측 (생존 + 폐업 전 구간 포함)
    - last_df: 가맹점별 지표 및 폐업여부 포함
    - 각 지표별 Prophet 학습 → 예측 → 임계치 비교 및 시각화
    """
    #기준년월(datetime) 변환
    df = last_df.copy()
    df['ds'] = pd.to_datetime(df['기준년월'], format='%Y%m')
    
    #  생존 / 폐업 데이터 분리
    alive_df = df[df['폐업여부'] == 0].copy()
    closed_df = df[df['폐업여부'] == 1].copy()

    #  폐업 매장: 폐업 전 pre_close_months개월만 포함
    # (매장별 기준년월이 있으므로 groupby로 처리)
    closed_pre = (
        closed_df.sort_values(['가맹점구분번호', 'ds'])
        .groupby('가맹점구분번호', group_keys=False)
        .apply(lambda g: g.tail(pre_close_months))
    )

    # 생존 + 폐업전 구간 통합
    total_df = pd.concat([alive_df, closed_pre], axis=0).reset_index(drop=True)
    total_df = total_df.sort_values(['가맹점구분번호', 'ds'])

    indicators = ['매출안정성지표', '경쟁우위 지표', '고객 충성도 지표']
    results = []

    # ==============================
    # Prophet 학습 및 예측 (지표별)
    # ==============================
    for target in indicators:
        sub = total_df[['ds', target]].dropna().copy()
        sub = sub.sort_values('ds')
        n = len(sub)
        if n < 10:
            continue

        prophet_df = sub.rename(columns={'ds': 'ds', target: 'y'})

        # Prophet 모델 생성 및 학습
        m = Prophet(
            yearly_seasonality=True,
            weekly_seasonality=False,
            daily_seasonality=False,
            changepoint_prior_scale=1
        )
      
        m.fit(prophet_df)

        # 미래 예측
        future = m.make_future_dataframe(periods=forecast_months, freq='MS')
        forecast = m.predict(future)

        # 성능 평가
        y_true = prophet_df['y'].iloc[-min(forecast_months, len(prophet_df)):]
        y_pred = forecast['yhat'].iloc[-min(forecast_months, len(forecast)):]
        mae, rmse, mape = evaluate_forecast(y_true, y_pred)

        warn_th = threshold_df.loc[target, '경고임계치']
        danger_th = threshold_df.loc[target, '위험임계치']

        # 시각화
        plt.figure(figsize=(10, 5))
       
        plt.plot(forecast['ds'], forecast['yhat'], color='#1f77b4', label='예측 추세')
        plt.axhline(y=warn_th, color='orange', linestyle='--', label=f'경고 {warn_th:.3f}')
        plt.axhline(y=danger_th, color='red', linestyle='--', label=f'위험 {danger_th:.3f}')
        plt.axvspan(forecast['ds'].iloc[-forecast_months], forecast['ds'].iloc[-1], 
                    color='khaki', alpha=0.2)
        plt.title(f'📈 {target} Prophet 예측 ')
        plt.xlabel('기준년월'); plt.ylabel(target)
        plt.legend(); plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()
        # 결과 저장
        results.append({
            '모델': 'Prophet',
            '지표': target,
            
            '예측 평균': y_pred.mean(),
            'MAE': mae,
            'RMSE': rmse,
            'MAPE(%)': mape,
            '경고임계치': warn_th,
            '위험임계치': danger_th
        })

    return pd.DataFrame(results)

```

```python
result_prophet = evaluate_forecast_model_prophet(
    last_df=last_df,
    threshold_df=threshold_df,
    forecast_months=10,   # 예측 기간
    pre_close_months=6   # 폐업 전 6개월만 포함
)

display(result_prophet)
```

```python
# ================================================
# 📦 가맹점별 맞춤형 Prophet 조기경보 시스템 (상위 10개만 시각화)
# ================================================
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------
# ⚙️ 1️⃣ 평가 지표 함수
# --------------------------------
def evaluate_forecast(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / np.clip(np.abs(y_true), 1e-8, None))) * 100
    return mae, rmse, mape


# --------------------------------
# ⚙️ 2️⃣ Prophet 기반 시계열 예측 함수 (가맹점별)
# --------------------------------
def evaluate_forecast_model_prophet_storewise(last_df, threshold_df, forecast_months=6, pre_close_months=6, plot_limit=4):
    """
    Prophet 기반 가맹점별 시계열 예측 + 맞춤형 임계선 + 경고 감지
    - last_df: 가맹점별 지표 데이터 (폐업여부 포함)
    - threshold_df: 지표별 임계치 데이터프레임
    - forecast_months: 예측 개월 수
    - pre_close_months: 폐업 매장은 폐업 전 몇 개월 데이터만 포함
    - plot_limit: 그래프 표시할 가맹점 개수 (기본 10)
    """

    # 날짜형 변환
    df = last_df.copy()
    df['ds'] = pd.to_datetime(df['기준년월'], format='%Y%m')

    # 생존 / 폐업 데이터 분리
    alive_df = df[df['폐업여부'] == 0].copy()
    closed_df = df[df['폐업여부'] == 1].copy()

    # 폐업 매장: 폐업 전 N개월만 포함
    closed_pre = (
        closed_df.sort_values(['가맹점구분번호', 'ds'])
        .groupby('가맹점구분번호', group_keys=False)
        .apply(lambda g: g.tail(pre_close_months))
    )

    # 생존 + 폐업전 구간 통합
    total_df = pd.concat([alive_df, closed_pre], axis=0).reset_index(drop=True)
    total_df = total_df.sort_values(['가맹점구분번호', 'ds'])

    # 예측할 지표
    indicators = ['매출안정성지표', '경쟁우위 지표', '고객 충성도 지표']

    # 결과 저장용 리스트
    all_results = []

    # --------------------------------
    # 가맹점별 Prophet 예측
    # --------------------------------
    store_ids = total_df['가맹점구분번호'].unique()[:4]
    for i, store_id in enumerate(store_ids):
        store_df = total_df[total_df['가맹점구분번호'] == store_id].copy()

        # 전체 평균 대비 비율 계산 (맞춤형 임계선용)
        global_mean = total_df[indicators].mean()
        local_mean = store_df[indicators].mean()
        ratio = local_mean / global_mean

        print(f"\n📍 [{store_id}] 맞춤 비율 (ratio):")
        print(ratio.round(3))

        for target in indicators:
            sub = store_df[['ds', target]].dropna().copy()
            sub = sub.sort_values('ds')
            n = len(sub)
            if n < 10:
                continue

            prophet_df = sub.rename(columns={'ds': 'ds', target: 'y'})

            # Prophet 모델 생성 및 학습
            m = Prophet(
                yearly_seasonality=True,
                weekly_seasonality=False,
                daily_seasonality=False,
                changepoint_prior_scale=1
            )
            m.fit(prophet_df)

            # 미래 예측
            future = m.make_future_dataframe(periods=forecast_months, freq='MS')
            forecast = m.predict(future)

            # 성능 평가
            y_true = prophet_df['y'].iloc[-min(forecast_months, len(prophet_df)):]
            y_pred = forecast['yhat'].iloc[-min(forecast_months, len(forecast)):]
            mae, rmse, mape = evaluate_forecast(y_true, y_pred)

            # 맞춤형 임계선 계산
            warn_th = threshold_df.loc[target, '경고임계치'] * ratio[target]
            danger_th = threshold_df.loc[target, '위험임계치'] * ratio[target]

            # 경고/위험 상태 플래그 생성
            forecast['경고'] = forecast['yhat'] < warn_th
            forecast['위험'] = forecast['yhat'] < danger_th

            # 마지막 상태 판단
            last_state = "안전"
            if forecast['위험'].iloc[-1]:
                last_state = "위험"
            elif forecast['경고'].iloc[-1]:
                last_state = "경고"

            # -------------------------
            # 시각화 (상위 10개만 출력)
            # -------------------------
            if i < plot_limit:
                plt.figure(figsize=(9, 5))
                plt.plot(sub['ds'], sub[target], color='gray', label='실측 추세')
                plt.plot(forecast['ds'], forecast['yhat'], color='#1f77b4', label='예측 추세')
                plt.axhline(y=warn_th, color='orange', linestyle='--', label=f'경고 {warn_th:.3f}')
                plt.axhline(y=danger_th, color='red', linestyle='--', label=f'위험 {danger_th:.3f}')
                plt.axvspan(forecast['ds'].iloc[-forecast_months], forecast['ds'].iloc[-1], 
                            color='khaki', alpha=0.2)
                plt.title(f'📈 [{store_id}] {target} 예측 (상태: {last_state})')
                plt.xlabel('기준년월'); plt.ylabel(target)
                plt.legend(); plt.grid(alpha=0.3)
                plt.tight_layout()
                plt.show()

            # -------------------------
            # 결과 저장
            # -------------------------
            all_results.append({
                '가맹점': store_id,
                '지표': target,
                '예측 평균': y_pred.mean(),
                'MAE': mae,
                'RMSE': rmse,
                'MAPE(%)': mape,
                '경고임계치(맞춤)': warn_th,
                '위험임계치(맞춤)': danger_th,
                '현재상태': last_state
            })

    result_df = pd.DataFrame(all_results)
    return result_df


# --------------------------------
# ⚙️ 3️⃣ 실행 예시
# --------------------------------
result_prophet_storewise = evaluate_forecast_model_prophet_storewise(last_df, threshold_df, plot_limit=4)

print("\n=== 📊 가맹점별 Prophet 예측 결과 요약 ===")
display(result_prophet_storewise)
```
